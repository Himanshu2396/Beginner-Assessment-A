```python
#Drop rows where Order ID or Product is missing.
df = df.dropna(subset=['Order ID', 'Product'])
```

```python
#Fill missing Price values with the mean price of that specific product.
df['Price Each'] = df['Price Each'].fillna(df.groupby('Product')['Price Each'].transform('mean'))
```

```python
#converting data type of order date column and 
df['Order Date'] = pd.to_datetime(df['Order Date'])
```
